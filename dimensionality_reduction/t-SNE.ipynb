## Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
```

## Connect Data

```python
df = pd.read_csv("../data.csv")
df.head()
```

## Visualize t-SNE

```python
x = df.iloc[:, 1:] # extract only features (remove label) to apply dimentionality reductions
x = pd.DataFrame(StandardScaler().fit_transform(x), columns=x.columns) # standardize the data
x.head()
```

```python
# apply t-SNE
tSNE = TSNE(n_components=2)
x_tSNE = tSNE.fit_transform(x)

# visualize t-SNE
fig, ax = plt.subplots(figsize=(8,8))
ax.scatter(x_tSNE[:, 0], x_tSNE[:, 1], c=df['Bankrupt?'].astype('category').cat.codes, alpha=0.5)
ax.set_title('t-SNE')
plt.show()
```

```python
## try 3D t-SNE

# apply t-SNE
tSNE = TSNE(n_components=3)
x_tSNE = tSNE.fit_transform(x)

# visualize t-SNE
#fig, ax = plt.subplots(figsize=(8,8))
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x_tSNE[:, 0], x_tSNE[:, 1], x_tSNE[:, 2], c=df['Bankrupt?'].astype('category').cat.codes, alpha=0.5)
ax.set_title('t-SNE')
plt.show()
```
